The dataset details are as follows:


*   overall - rating given a particular reviewer
*   reviewTime - Time of review
*   reviewerID - ID of the reviewer
*   style - depicts the clothing style
*   reviewText - review in text frmat
*   summary - description of the product
*   product_id - ID of the product
*   place - region of the reviwer
*   age - age of the product
*   gender - gender of the reviwer







```python
data = pd.read_csv('data.csv')
data = data.drop(['Unnamed: 0'], axis=1)
data
```

```python
data.isnull().sum()
```

Pivot the dataframe with columns as product_id, index as reviewerId and values as overall

```python
data_pivot = pd.pivot_table(data, index="reviewerID", columns="product_id", values="overall")
data_pivot = data_pivot.fillna(0)
data_pivot
```

## **Content Based Filtering**

Correlation matrix for the product_id's based on the ratings given by the reviewer


```python
data_pivot.corr()
```

## **Collaborative Filtering**

Demonstrating the algorithm with an example reviewer input


```python
user_input = "AZ7LUSY20ZFSC"
user_df = data_pivot[data_pivot.index == user_input]
user_df
```

extracting the other reviewers who purchased some of the products purchased by the input reviewer

```python
userpro = []
for i in user_df.columns:
  if user_df[i][0] != 0:
    userpro.append(i)
```

```python
consol_data = data_pivot[userpro]           #consolidated data
consol_data
```

Calculating the number of products in similar to the input reviewer with respect to other reviwerers

```python
l = []
for i, row in consol_data.iterrows():
    li = list(row)
    v = len(userpro)-li.count(0)
    l.append(v)

consol_data['total'] = l
consol_data
```

Reviewers with greater than 2 similar products

```python
users_same_products = consol_data[consol_data["total"] >= 2]["total"]
print(users_same_products.count())
```

```python
user_index = list(users_same_products.index)
```

Correlation matrix for the similar reviewers

```python
final_df = pd.concat([consol_data[consol_data.index.isin(user_index)],
                      user_df[userpro]])

final_df.shape
final_df[:-1].T.corr()
```

```python
corr_df = final_df[:-1].T.corr().unstack().sort_values().drop_duplicates()
corr_df = pd.DataFrame(corr_df, columns=["corr"])
corr_df.index.names = ['user_id_1', 'user_id_2']
corr_df = corr_df.reset_index()
corr_df.head()
```

Sorting the reviewers with respect to correlation values

```python
top_users = corr_df[(corr_df["user_id_1"] == user_input) & (corr_df["corr"] >= 0)][
    ["user_id_2", "corr"]].reset_index(drop=True)

top_users = top_users.sort_values(by='corr', ascending=False)
top_users.rename(columns={"user_id_2": "reviewerId"}, inplace=True)
top_users
```

```python
ref = list(top_users['reviewerId'])
ref
```

```python
d = data.copy()
dd = pd.DataFrame()
cor_list = []
for i in ref:
    df = d[d['reviewerID'] == i]
    c = 0
    for j in range(len(df)):
      df2 = top_users[top_users['reviewerId'] == i]
      cor_list.append(df2['corr'].values[0])
      c+=1

    dd = dd.append(df, ignore_index = True)
top_users_ratings = dd.drop(['reviewTime', 'style', 'reviewText',	'summary', 'place', 'age',	'gender'], axis=1)
top_users_ratings['corr'] = cor_list
top_users_ratings.head()
```

Calculating the aggregate rating between corr and it's overall  

```python
top_users_ratings['aggre_rating'] = top_users_ratings['corr'] * top_users_ratings['overall']
top_users_ratings.groupby('product_id').agg({"aggre_rating": "mean"})
top_users_ratings.head()
```

Product Recommendation by **Collaborative Filtering**

Top 5 recommendation based on Collaborative Filtering

```python
recommendation_df = top_users_ratings.groupby('product_id').agg({"aggre_rating": "mean"})
recommendation_df = recommendation_df.reset_index()
recommendation_df.sort_values("aggre_rating", ascending=False).head()
```

Fetching the similarity matrices for demographic features such as age, gender, place

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics.pairwise import cosine_similarity

demo_data = data.drop(['overall',	'reviewTime',	'style',	'reviewText',	'summary'], axis=1)

label_encoder = LabelEncoder()
demo_data['gender_encoded'] = label_encoder.fit_transform(demo_data['gender'])
demo_data['place_encoded'] = label_encoder.fit_transform(demo_data['place'])

place_product_matrix = demo_data.pivot_table(index='reviewerID', columns='product_id', values='place_encoded', fill_value=0)
place_similarity = place_product_matrix.corr()

gender_product_matrix = demo_data.pivot_table(index='reviewerID', columns='product_id', values='gender_encoded', fill_value=0)
gender_similarity = gender_product_matrix.corr()

age_product_matrix = demo_data.pivot_table(index='reviewerID', columns='product_id', values='age', fill_value=0)
age_similarity = age_product_matrix.corr()
```

Correlation matrix with respect to place

```python
place_similarity
```

Correlation matrix with respect to age

```python
age_similarity
```

Correlation matrix with respect to gender

```python
gender_similarity
```

Calculating a hybrid correlation matrix by combining all the 3 demgraphic feactures based correlation matrices

```python
user_similarity = 0.2*gender_similarity.corr() + 0.2*place_similarity.corr() + 0.6*age_similarity.corr()
user_similarity
```

Calculating complete correlation matrix by considering the demographic correlation and the content based (product-product) correlation matrix to fetch the top 5 recommendations

```python
user_similarity = 0.4*user_similarity + 0.6*data_pivot.corr()
user_similarity
```

```python
product_input = 'P09395'

corr_df = user_similarity[:-1].T.corr().unstack().sort_values().drop_duplicates()
corr_df = pd.DataFrame(corr_df, columns=["corr"])
corr_df.index.names = ['p_id_1', 'p_id_2']
corr_df = corr_df.reset_index()
corr_df.head()
```

Product Recommendation based on **Content and Demographic filtering**

Top 5 Products based on Content and Demographic filtering

```python
top_products = corr_df[(corr_df["p_id_1"] == product_input) & (corr_df["corr"] >= 0)][
    ["p_id_2", "corr"]].reset_index(drop=True)

top_products = top_products.sort_values(by='corr', ascending=False)
top_products.rename(columns={"p_id_2": "product_id"}, inplace=True)
top_products.head()
```
